## Limit Buy Order

```python
client.get_account()
```

```python
order1 = client.create_order(symbol = "BTCUSDT", side = "BUY", type = "LIMIT",
                             quantity = 0.0001, timeInForce = "GTC", price = 50000)
#비트코인이 50000달러로 내려갈 경우 0.0001의 비트코인을 매수한다는 뜻 
#timeInForce = "GTC" 주문을 얼마나 시행할지 정하는 것 보통 GTC를 넣는데 취소시한 유효주문의 줄임말이다
#주문이 체결되거나 취소할 때까지 유지한다는 뜻이다.
```

```python
order1
```

```python
order_id_buy = order1["orderId"]
order_id_buy
```

```python
client.get_order(symbol="BTCUSDT", orderId = order_id_buy)
```

## Limit Sell Order

```python
order2 = client.create_order(symbol = "BTCUSDT", side = "SELL", type = "LIMIT",
                             quantity = 0.0001, timeInForce = "GTC", price = 60000)
```

```python
order2
```

```python
order_id_sell = order2["orderId"]
order_id_sell
```

```python
#get_open_orders() 메소들르 사용하여 특정 조합의 미체결 주문을 전부 가져올 수 있다.
open_orders = client.get_open_orders(symbol="BTCUSDT")
open_orders
```

```python
pd.DataFrame(open_orders)
```

```python
#cancel_order()메소드를 통하여 특정 orderId의 미체결 주문을 취소할 수 있다.
result = client.cancel_order(symbol = "BTCUSDT", orderId = order_id_buy)
result
```

```python
result = client.cancel_order(symbol = "BTCUSDT", orderId = order_id_sell)
result
```

```python
#get_open_orders() 다시확인했을때 주문들이 다 취소되어있는 것을 확인할 수 있다.
open_orders = client.get_open_orders(symbol="BTCUSDT")
open_orders
```
