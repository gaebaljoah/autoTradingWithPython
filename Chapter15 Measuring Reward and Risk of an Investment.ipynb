```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
plt.style.use("seaborn-v0_8")
```

```python
msft = pd.read_csv("msft.csv", index_col = "Date", parse_dates = ["Date"])
msft
```

```python
msft.Price.plot(figsize = (15,8), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

```python
msft.describe()
#count: 데이터의 총 개수
#mean: 평균값
#std: 표준편차
#min: 최솟값
#25%, 50%, 75%: 데이터의 백분위수 (25%, 50%, 75%에 해당하는 값)
#max: 최댓값
```

```python
mu = msft.Returns.mean() #일일 산술평균의 수익률 
#산술평균은 일반적으로 "평균"이라고도 불리는 통계적 개념 데이터 집합의 값들의 합을 총 개수로 나눈 값으로, 데이터 집합의 대푯값을 나타낸다
mu
```

```python
sigma = msft.Returns.std() #일일 수익률의 표준편차
sigma
```

```python
np.sqrt(msft.Returns.var()) #일일 수익률의 표준편차 sqrt() 메서드 활용
```
