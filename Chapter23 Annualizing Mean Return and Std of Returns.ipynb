```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.6f}'.format
plt.style.use("seaborn-v0_8")
```

```python
msft = pd.read_csv("msft.csv", index_col="Date", parse_dates = ["Date"])
```

```python
msft
```

```python
msft["log_ret"] = np.log(msft.Price / msft.Price.shift())
msft
```

```python
msft.log_ret.agg(["mean","std"])
#평균 수익률 0.1%
#표준편차 1.7%
```

```python
#로그수익률의 평균을 연으로 곱하여 연간 기대수익률을 나타낼 수 있다.
ann_mu = msft.log_ret.mean() * 252
ann_mu
```

```python
#연환산 평균 수익률(ann_mu)를 활용하여 연평균 성장률을 산출할 수도 있다.
carg = np.exp(ann_mu) -1
carg
```

```python
ann_std = msft.log_ret.std() * np.sqrt(252) 
#sqrt()메서드를 활용하여 일일 평균 수익률의 기준편차를 연환산화 할 수 있다.
#연간 수익률의 표준편차가 높을수록 해당 주식의 가격이 연간으로 얼마나 변동하는지를 나타내며, 
#이는 투자의 위험 수준을 평가하는 데 중요한 지표이다.
ann_std
```

### Resampling/Smoothing

```python
msft.head(25)
```

```python
msft.Price.plot(figsize = (12,8))
plt.legend()
plt.show()
```

```python
monthly = msft.Price.resample("M").last()
monthly
```

```python
monthly.plot(figsize = (12, 8))
plt.legend()
plt.show()
```

```python
freqs = ["A","Q","M","W-Fri","D"]
periods = [1, 4, 12, 52, 252]
ann_mean = []
ann_std = []
```

```python
for i in range(5):
    resamp = msft.Price.resample(freqs[i]).last()
    ann_mean.append(np.log(resamp / resamp.shift()).mean() * periods[i])
    ann_std.append(np.log(resamp / resamp.shift()).std() * np.sqrt(periods[i]))
```

```python
ann_mean
```

```python
ann_std
```

```python
summary = pd.DataFrame(data = {"ann_std":ann_std, "ann_mean":ann_mean}, index = freqs)
summary
```

```python
summary.plot(kind = "scatter", x = "ann_std", y = "ann_mean", figsize = (15,12), s = 50, fontsize = 15)
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, "ann_std"]+0.001, summary.loc[i,"ann_mean"]+0.001), size =15)
plt.ylim(0, 0.3)
plt.xlabel("ann. Risk(std)", fontsize = 15)
plt.ylabel("ann. Return",fontsize = 15)
plt.title("Risk/Return", fontsize = 20)
plt.show()
```

### Rolling Statistics
롤링 통계들은 재무와 투자에서의 근본적인 규칙 분석하고 증명하는 것에 도움을 준다.
과거 성과가 미래에 나올 성과의 최적의 지표는 아니라는 것이다.
미래 성과를 예측할때 과거 성과를 고려해야 한다는 관점이 있는데, 관점일뿐이고 과거에 높은 수익을 낸 상품이나 전략이 미래에도 계속해서 그럴 것이라고 가정할 순 없는것이다.

```python
msft
```

```python
#연환산 평균 수익률
ann_mu = msft.log_ret.mean() * 252
ann_mu
```

```python
#연환산 수익률 표준편차(연환산 위험도)
ann_std = msft.log_ret.std() * np.sqrt(252)
ann_std
```

수익률과 위험성은 시간에 따라 변화하고 롤링 통계로 측정하고 수량화가 가능하다.

```python
window = 252
```

```python
msft.log_ret.rolling(window = 252)
```

```python
msft.log_ret.rolling(window = 252).sum()
```

```python
roll_mean = msft.log_ret.rolling(window = 252).mean() * 252
roll_mean
```

```python
roll_mean.iloc[250:]
```

```python
roll_mean.plot(figsize = (12,8))
plt.show()
```

```python
roll_std = msft.log_ret.rolling(window = 252).std() * np.sqrt(252)
roll_std
```

```python
roll_std.plot(figsize = (12,8))
plt.show()
```

```python
roll_mean.plot(figsize=(12, 8))
roll_std.plot()
plt.show()
#파란 선이 rolling 평균수익률 
#초록 선은 수익률의 롤링 표준 편차 
```
