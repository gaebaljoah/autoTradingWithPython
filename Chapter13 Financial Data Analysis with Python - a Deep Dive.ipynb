### **Normalizing Financial Time Series to a Base Value (100)**
###### 절대 가격으로는 금융 상품의 실적을 비교할 수 없다
###### 한가지 해결책은 가격을 정규화 하는 것인데 예를 들어 100이라는 범위를 정해줄 수 있다.

```python
close
```

```python
close.iloc[0,0] #first Price AMC
```

```python
close.BA.div(close.iloc[0,0])
```

```python
close.iloc[0]
```

```python
norm = close.div(close.iloc[0]).mul(100)
norm
```

```python
norm.dropna().plot(figsize = (15, 8), fontsize = 13, logy = True)
plt.legend(fontsize = 13)
plt.show()
```
